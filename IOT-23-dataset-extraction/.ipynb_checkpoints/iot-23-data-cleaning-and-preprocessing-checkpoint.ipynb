# Supervised ML methods for anomaly detection in IOT to enahnce network security
## Part 1 - DATA CLEANING AND PREPROCESSING

The IoT-23 dataset is a collection of network traffic from Internet of Things (IoT) devices. It includes 20 malware captures executed in IoT devices, and 3 hotspot captures for benign IoT devices traffic12. The 3 hotspot captures are not being included in the data cleaning because this feature was not considered relevant for the specific analysis being performed.

In this notebook, we load the raw dataset file and implement initial cleaning to prepare it for the next processing phase.

> **INPUT:** downloaded the raw dataset file from its original source. <br>
> **OUTPUT:** a cleaned version of the dataset stored to an intermediate csv file.

***

### 1. INITIALIZATION

```python
# Import necessary libraries and modules
import pandas as pd
import numpy as np
import glob
import os
import ipaddress
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.impute import KNNImputer
```

```python
# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('future.no_silent_downcasting', True)
```

```python
# Initialize required variables to read from the data file
for dirname, _, filenames in os.walk('../CSV-data/raw/'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

### 2. LOADING DATASET FILE (*.labeled) AND CONVERT TO CSV FILE

```python
# Get a list of all the .labeled files
labeled_files = glob.glob('../CSV-data/raw/*.labeled')

# Create an empty list to hold dataframes
dfs = []

# Loop over the list of .labeled files
for labeled_file in labeled_files:    
    # Column names are included in the file in a commented line, so we need to read the corresponding line separately and remove the first description word.
    data_columns = pd.read_csv(labeled_file, sep='\t', skiprows=6, nrows=1, header=None).iloc[0][1:]
    
    # Read the actual dataset
    data_df = pd.read_csv(labeled_file, sep='\t', nrows=100000, comment="#", header=None)
    
    # Set column names
    data_df.columns = data_columns
    
    # Read each .labeled file into a dataframe and append to the list
    dfs.append(data_df)

# Concatenate all dataframes in the list
combined_df = pd.concat(dfs, ignore_index=True)

# Write the combined dataframe to a new csv file
combined_df.to_csv('../CSV-data/interim/iot23_combined.csv', index=False)
```

### 3. FIX COMBINED COLUMNS

The last column in the dataset contains three separate values and needs to be unpacked into three corresponding columns. This is due to unmatched separators in the original data file.

```python
# Split the last combined column into three ones
tunnel_parents_column = combined_df.iloc[:,-1].apply(lambda x: x.split()[0])
label_column = combined_df.iloc[:,-1].apply(lambda x: x.split()[1])
detailed_label_column = combined_df.iloc[:,-1].apply(lambda x: x.split()[2])
```

```python
# Check the dataset
combined_df.head()
```

```python
# Drop the combined column
combined_df.drop(["tunnel_parents   label   detailed-label"], axis=1, inplace=True)
```

```python
# Add newly created columns to the dataset
combined_df["tunnel_parents"] = tunnel_parents_column
combined_df["label"] = label_column
combined_df["detailed_label"] = detailed_label_column
```

```python
# Check the dataset
combined_df.head()
```

```python
combined_df['label'].value_counts()
```

```python
combined_df['detailed_label'].value_counts()
```

### 4. DROPPING OF NON-UNIQUE AND UNIMPORTANT COLUMNS

```python
# Check the number of unique values in each column
combined_df.nunique().sort_values(ascending=False)
```

```python
# Replace all occurrences of empty/unset cells with null values
combined_df.replace({'-':np.nan, "(empty)":np.nan}, inplace=True)
```

```python
# Fix data types of the misinterpreted columns
dtype_convert_dict = {
    "duration": float,
    "orig_bytes": float,
    "resp_bytes": float
}
combined_df = combined_df.astype(dtype_convert_dict)
```

```python
# Drop columns that are non-unique, which do not contribute to model building
combined_df.drop(columns=["ts","uid","local_resp","local_orig","tunnel_parents","detailed_label"], inplace=True)
```

```python
# Check the dataset
combined_df.head()
```

### 6. CONVERTING CATEGORICAL COLUMNS TO NUMERICAL DATATYPE

```python
combined_df.head()
```

```python
# A method to encode an ip address using the IPv4/IPv6 manipulation library
def encode_ipv4(ip):
    return int(ipaddress.IPv4Address(ip))
```

```python
# Encode ip features into numerical values
combined_df["id.orig_h"] = combined_df["id.orig_h"].apply(encode_ipv4)
combined_df["id.resp_h"] = combined_df["id.resp_h"].apply(encode_ipv4)
```

```python
# Check encoded ip addresses
ip_columns = ["id.orig_h","id.resp_h"]
for c in ip_columns:
    print(f"IP col ({c}) values: {combined_df[c].unique()}")
```

```python
# Initialize encoder with default parameters
target_le = LabelEncoder()

# Fit the encoder to the target attribute
encoded_attribute = target_le.fit_transform(combined_df["label"])

# Replace target attribute with encoded values
combined_df["label"] = encoded_attribute

# Check mapped labels
dict(zip(target_le.classes_, target_le.transform(target_le.classes_)))
```

```python
# Fix data types of the misinterpreted columns
dtype_convert_dict = {
    "label": float
}
combined_df = combined_df.astype(dtype_convert_dict)
```

```python
#Impute categorical missing values
columns_with_missing = ["service","history"]

# Initialize the imputer with 'most_frequent' strategy
imp = SimpleImputer(strategy="most_frequent")

# Fit and transform the dataset
combined_df[columns_with_missing] = imp.fit_transform(combined_df[columns_with_missing])

# Now df contains imputed values for "service" and "history"
combined_df.head()
```

```python
# Check the number of unique values in each feature
categorical_features = ["proto","service","conn_state","history"]
for c in categorical_features:
    print(f"Column ({c}) has ({combined_df[c].nunique()}) distinct values.")
```

```python
combined_df.head()
```

```python
# Check values of "history" because it has too many unique values.
history_val_counts = combined_df["history"].value_counts()
history_val_counts
```

```python
# Map values to their frequencies
history_freq_map = combined_df["history"].map(history_val_counts)

# Replace low frequent values in "history" with "Other" using the corresponding frequency map
combined_df["history"] = combined_df["history"].mask(history_freq_map < 100, "Other")

# Check "history" values after mapping
combined_df["history"].value_counts()
```

### 7. CONVERTING MULTI-CLASS LABELS TO BINARY LABELS

```python
# Initialize the encoder with its default parameters
ohe = OneHotEncoder()

# Fit the encoder to categorical features in the dataset
encoded_features = ohe.fit_transform(combined_df[categorical_features])

# Create a dataframe of encoded features
encoded_features_df = pd.DataFrame(encoded_features.toarray(), columns=ohe.get_feature_names_out())

# Check the results of encoding
encoded_features_df
```

```python
# Merge encoded features with the dataset and drop original columns
combined_df = pd.concat([combined_df, encoded_features_df], axis=1).drop(categorical_features, axis=1)
```

```python
combined_df.head()
```

### 8. HANDLING MISSING VALUES

```python
# Check the number of null values in each column
combined_df.isnull().sum().sort_values(ascending=False)
```

```python
# def introduce_nan(x,percentage):
#     n = int(len(x)*(percentage - x.isna().mean()))
#     idxs = np.random.choice(len(x), max(n,0), replace=False, p=x.notna()/x.notna().sum())
#     x.iloc[idxs] = np.nan

# combined_df.apply(introduce_nan, percentage=.5)
# combined_df.info()
```

```python
# Replace missing values in 'duration', 'orig_bytes', and 'resp_bytes' with the mean of their respective columns or remove them, depending on you.
columns = ['duration', 'orig_bytes', 'resp_bytes']

# Replace missing values with the mean of their respective columns
for col in columns:
    combined_df[col] = combined_df[col].replace(np.nan, combined_df[col].mean())
```

```python
combined_df.head()
```

```python
# Check the number of null values in each column
combined_df.isnull().sum().sort_values(ascending=False)
```

### 9. SCALING NUMERICAL ATTRIBUTES

As we aim to compare several classifiers, and since some of them rely on distance-based comparisons, we will scale the numerical features to have them represented in a unified distribution.

```python
# Check statistics for numerical features
numerical_features = ["id.orig_h","id.orig_p","id.resp_h","id.resp_p","duration","orig_bytes","resp_bytes","missed_bytes","orig_pkts","orig_ip_bytes","resp_pkts","resp_ip_bytes"]
pd.set_option("display.float", "{:.2f}".format)
combined_df[numerical_features].describe()
```

Since most attributes have no normal distribution, it's more suitable to apply a normalization (between 0 and 1, using MinMaxScaler) instead of standardization.

```python
# Initialize and apply MinMaxScaler scaler
min_max_scaler = MinMaxScaler()
combined_df[numerical_features] = min_max_scaler.fit_transform(combined_df[numerical_features])
```

```python
# Check statistics for scaled features
combined_df[numerical_features].describe()
```
